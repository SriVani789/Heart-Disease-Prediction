```python
import pandas as pd
```

```python
data=pd.read_csv('heart.csv')
```

```python
data
```

```python
inputs=data[["Age","Sex","ChestPainType","RestingBP","Cholesterol","FastingBS","RestingECG","MaxHR","ExerciseAngina","Oldpeak","ST_Slope"]]
inputs
```

```python
targets=data["HeartDisease"]
targets
```

```python
inputs.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
inputs['Age_n'] = LabelEncoder().fit_transform(inputs['Age'])
inputs['Sex_n'] = LabelEncoder().fit_transform(inputs['Sex'])
inputs['ChestPainType_n'] = LabelEncoder().fit_transform(inputs['ChestPainType'])
inputs['RestingBP_n'] = LabelEncoder().fit_transform(inputs['RestingBP'])
inputs['Cholesterol_n'] = LabelEncoder().fit_transform(inputs['Cholesterol'])
inputs['FastingBS_n'] = LabelEncoder().fit_transform(inputs['FastingBS'])
inputs['RestingECG_n'] = LabelEncoder().fit_transform(inputs['RestingECG'])
inputs['MaxHR_n'] = LabelEncoder().fit_transform(inputs['MaxHR'])
inputs['ExerciseAngina_n'] = LabelEncoder().fit_transform(inputs['ExerciseAngina'])
inputs['Oldpeak_n'] = LabelEncoder().fit_transform(inputs['Oldpeak'])
inputs['ST_Slope_n'] = LabelEncoder().fit_transform(inputs['ST_Slope'])
```

```python
inputs
```

```python
inputs_n=inputs.drop(["Age","Sex","ChestPainType","RestingBP","Cholesterol","FastingBS","RestingECG","MaxHR","ExerciseAngina","Oldpeak","ST_Slope"],axis="columns")
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.linear_model import LogisticRegression

# Instantiate the model
model = LogisticRegression()

# Fit the model
model.fit(inputs_n, targets)
```

```python
model.score(inputs_n,targets)
```

```python
sample_input = [[0, 1, 2, 3, 4, 5, 0, 6, 1, 7, 2]]  # Replace with the actual values
predicted_class = model.predict(sample_input)
print(predicted_class)
```

```python
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# Make predictions
predicted = model.predict(inputs_n)

# Calculate the confusion matrix
cm = confusion_matrix(targets, predicted)

# Extract true positive, true negative, false positive, and false negative values
tn, fp, fn, tp = cm.ravel()

# Display the confusion matrix
sns.set(font_scale=1.2)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Heart Disease", "Heart Disease"], yticklabels=["No Heart Disease", "Heart Disease"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

# Print true positive, true negative, false positive, and false negative
print(f"True Positives: {tp}")
print(f"True Negatives: {tn}")
print(f"False Positives: {fp}")
print(f"False Negatives: {fn}")
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Instantiate and fit a Decision Tree Classifier
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(inputs_n, targets)

# Plot the decision tree
plt.figure(figsize=(15, 10))
plot_tree(decision_tree_model, feature_names=inputs_n.columns, class_names=["No Heart Disease", "Heart Disease"], filled=True, rounded=True)
plt.show()
```

```python
report = classification_report(targets, predicted, target_names=["No Heart Disease", "Heart Disease"])
print(report)
```
